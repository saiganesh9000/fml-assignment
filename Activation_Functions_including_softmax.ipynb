```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Sigmoid Function**

```python
def sigmoid(x):
  return 1/(1 + np.exp(-x))

x = np.linspace(-10 , 10)
plt.plot(x , sigmoid(x))
plt.axis('tight')
plt.title('Activation Fuction : Sigmoid')
plt.show()
```

# **tanh**

```python
def tanh(x):
  return np.tanh(x)

x = np.linspace(-10 , 10)
plt.plot(x , sigmoid(x))
plt.axis('tight')
plt.title('Activation Fuction : tanh(x)')
plt.show()
```

# **ReLU**

```python
def RELU(x):
  x1 = []
  for i in x :
    if i < 0 :
      x1.append(0)
    else:
      x1.append(i)
  return x1


x = np.linspace(-10 , 10)
plt.plot(x , RELU(x))
plt.axis('tight')
plt.title('Activation Fuction : RELU')
plt.show()
```

# **Softmax**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(x):
  return np.exp(x)/np.sum(np.exp(x) , axis = 0 )

x = np.linspace(-10 , 10)
plt.plot(x , softmax(x))
plt.axis('tight')
plt.title('Activation Function : Softmax')
plt.show()
```
